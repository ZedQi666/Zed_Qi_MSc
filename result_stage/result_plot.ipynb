```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
import json 
bert_train = []
bert_test = []
for filename in os.listdir("bert"):
    print(filename)
    filepath = os.path.join("bert", filename)
    with open(filepath,'r', encoding='UTF-8') as f:
         load_dict = json.load(f)
    bert_train.append(load_dict['training_accuracy'])
    bert_test.append(load_dict['validation_accuracy'])
```

```python
bert_gru_train = []
bert_gru_test = []
for filename in os.listdir("bert+gru"):
    print(filename)
    filepath = os.path.join("bert+gru", filename)
    with open(filepath,'r', encoding='UTF-8') as f:
         load_dict = json.load(f)
    bert_gru_train.append(load_dict['training_accuracy'])
    bert_gru_test.append(load_dict['validation_accuracy'])
```

```python
plt.figure(figsize=(10,6))
plt.plot(bert_gru_train, label="Bert+GRU")
plt.plot(bert_train, label="Bert")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Train Dataset")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(bert_gru_test, label="Bert+GRU")
plt.plot(bert_test, label="Bert")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Validation Dataset")
plt.show()
```
